# PyCity Schools Analysis

* In general, schools with smaller student populations outperformed schools with larger student populations. The threshold is around 2000 students. Schools with under 2000 students did better on the exams than schools with over 2000 students. 

* Scores were consistent across all grades in each school. For example, 12th graders did not outperform 9th graders at individual schools, so a students' grade or level doesn't predict performance. 

* An unexpected outcome is that spending more per student doesn't increase performance. Schools with lower per student spending did better than schools that spend more per student. 
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score 
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count)*100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Number of Schools": [15],
    "Total Students": [39170],
    "Total Budget" : [24649428],
    "Average Math Score": [78.98],
    "Average Reading Score":[81.87],
    "Math Passed %":[74.98],
    "Reading Passed %": [85.80],
    "Overall Passing %": [65.17]
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data.set_index(["school_name"])["type"]
# school_types
```

```python
# Calculate the total student count
per_school_counts = school_data_complete.groupby(["school_name"]).count()["Student ID"]
# per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
# per_school_budget
```

```python
# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
# per_school_math
```

```python
# Calculate the number of schools with math scores of 70 or higher
# Interpretation of above Q - Calculate the total number of students who passed math at each school

# school_passing_math = len(school_data_complete[(school_data_complete["math_score"] >= 70)].groupby(["school_name"]))
school_passing_math = (school_data_complete[(school_data_complete["math_score"] >= 70)])
# school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = (school_data_complete[(school_data_complete["reading_score"] >= 70)])
# school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
# passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
# per_school_passing_reading
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.DataFrame(columns=['School Type','Total Students', 'Total School Budget', 'Per Student Budget', 
                                           'Per School Math', 'Per School Reading'])
per_school_summary['School Type'] = school_types
per_school_summary['Total Students'] = per_school_counts
per_school_summary['Total School Budget'] = per_school_budget
per_school_summary['Per Student Budget'] = per_school_capita
per_school_summary['Per School Math'] = per_school_passing_math
per_school_summary['Per School Reading'] = per_school_passing_reading
per_school_summary['Per School Summary'] = overall_passing_rate

# Formatting
# per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,.0f}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
# per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
# per_school_summary["Per School Math"] = per_school_summary["Per School Math"].map("{:.2f}".format)
# per_school_summary["Per School Reading"] = per_school_summary["Per School Reading"].map("{:.2f}".format)
# per_school_summary["Per School Summary"] = per_school_summary["Per School Summary"].map("{:.2f}".format)

# Display the DataFrame
per_school_summary
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
# Use the code to select only the `math_score`.
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()['math_score']
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()['math_score']
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()['math_score']
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()['math_score']

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame(columns=['9th Grade Math Average', '10th Grade Math Average', 
                                             '11th Grade Math Average', '12th Grade Math Average'])

math_scores_by_grade['9th Grade Math Average'] = ninth_graders_scores
math_scores_by_grade['10th Grade Math Average'] = tenth_graders_scores
math_scores_by_grade['11th Grade Math Average'] = eleventh_graders_scores
math_scores_by_grade['12th Grade Math Average'] = twelfth_graders_scores

# Formatting
math_scores_by_grade['9th Grade Math Average'] = math_scores_by_grade['9th Grade Math Average'].map("{:.2f}".format)
math_scores_by_grade['10th Grade Math Average'] = math_scores_by_grade['10th Grade Math Average'].map("{:.2f}".format)
math_scores_by_grade['11th Grade Math Average'] = math_scores_by_grade['11th Grade Math Average'].map("{:.2f}".format)
math_scores_by_grade['12th Grade Math Average'] = math_scores_by_grade['12th Grade Math Average'].map("{:.2f}".format)

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Scores by School Spending

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
#sort Per Student Budget column
school_spending_df.sort_values(by=['Per Student Budget'], ascending=True)
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["$0-585", "$585-630", "$630-645", "$645-680"]
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'], spending_bins, labels=labels, include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()['Per School Math']
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()['Per School Reading']
spending_total_students = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()['Total Students']
# spending_passing_math = school_data_complete.groupby(["Spending Ranges (Per Student)"]).mean()[passing_math_percentage]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).sum()['Per School Math'] / spending_total_students * 100
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).sum()['Per School Reading'] / spending_total_students * 100
spending_passing_overall = school_spending_df.groupby(["Spending Ranges (Per Student)"]).sum()['Per School Summary'] / spending_total_students * 100

```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame(columns=['Average Math Score','Average Reading Score', '% Passing Math', '% Passing Reading', 
                                           '% Overall Passing'])
                                         
spending_summary['Average Math Score'] = spending_math_scores
spending_summary['Average Reading Score'] = spending_reading_scores
spending_summary['% Passing Math'] = spending_passing_math
spending_summary['% Passing Reading'] = spending_passing_reading
spending_summary['% Overall Passing'] = spending_passing_overall

# Display results
spending_summary
```
